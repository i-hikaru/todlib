How to use TOD class (todlib.py)
================================

TOD クラスは時系列データを扱うクラスです。使用する際は、`ahiru:/home/hikaru/public_html/status/todlib.py` にあるものをコピーして使うか、

    $ git clone https://github.com/i-hikaru/todlib.git
    
としてください。folk は歓迎します。

TOD クラスは以下のメソッドを持ちます。

<dl>
<dt>todlib.ts(skip=1)</dt>
<dd>TOD のタイムスタンプを返すメソッド。</dd>
<pre>
Parameters: skip: int
            指定した整数だけデータを間引きます

Returns: out: numpy.ndarray
</pre>
</dl>

<dl>
<dt>todlib.dates(skip=1, tz=pytz.timezone('Asia/Tokyo'))</dt>
<dd>TOD の時間情報を datetime モジュールの形式で返すメソッド。</dd>
<pre>
Parameters: skip: int
            指定した整数だけデータを間引きます
            tz: pytz.tzfile
            pytz モジュールのタイムゾーンを指定します
            
Returns: out: numpy.ndarray
</pre>
</dl>

<dl>
<dt>todlib.vals(skip=1)</dt>
<dd>TOD のすべての値を返すメソッド。</dd>
<pre>
Parameters: skip: int
            指定した整数だけデータを間引きます

Returns: out: numpy.ndarray
</pre>
</dl>

<dl>
<dt>todlib.get_channel_vals(channel, skip=1)</dt>
<dd>TOD の各チャンネルの値を返すメソッド。</dd>
<pre>
Parameters: channel: int
            チャンネル番号を指定します
            skip: int
            指定した整数だけデータを間引きます

Returns: out: numpy.ndarray
</pre>
</dl>

<dl>
<dt>todlib.gentod(skip=1, colslabel=None)</dt>
<dd>TOD を pandas モジュールの DataFrame 形式で返すメソッド。</dd>
<pre>
Parameters: skip: int
            指定した整数だけデータを間引きます
            colslabel: list
            各チャンネルをラベリングします

Returns: out: Pandas.core.frame.DataFrame
</pre>
</dl>

<dl>
<dt>todlib.gentod_all(direcroty=None, extension='dat', colslabel=None)</dt>
<dd>イニシャライズしたファイルの属するディレクトリにあるすべての TOD を pandas モジュールの DataFrame 形式で返すメソッド。</dd>
<pre>
Parameters: direcroty: str
            データファイルが属するディレクトリを指定できます
            extension: str
            データファイルの拡張子を指定できます
            colslabel: list
            各チャンネルをラベリングします

Returns: out: Pandas.core.frame.DataFrame
</pre>
</dl>


Notes
-----

TOD クラスは使用するデータのある path と使用する列を引数にとってオブジェクトをつくります。ここでは、LakeShore 218 で取得した 8 チャンネルの温度データを例に TOD クラスの使い方を説明します。


Example
-------

LakeShore の温度データ（`/home/gb/public_html/gbmonitor/temp/data/lastest`）は 10 列の値（時刻, タイムスタンプ, チャンネル 1 の温度, ..., チャンネル 8 の温度）からなります。TOD クラスは 1 列目にタイムスタンプ、2 列目以降をそのときの値（この場合は温度）となるようにデータを指定します。いまの場合は 1 列目以降の列を `usecols` に指定します。

```python
%matplotlib inline
import matplotlib.pyplot as plt
import matplotlib.dates as md
import pytz
import pandas as pd
import todlib

tod = todlib.TOD('/home/gb/public_html/gbmonitor/temp/data/lastest', usecols=range(1, 10))
```

各チャンネルの全期間の温度をプロットするには以下のようにします。

```python
fig, ax = plt.subplots()
for i in range(8):
    ax.plot(tod.dates(10), tod.vals(10)[:,i], label='ch. '+str(i))
ax.xaxis.set_major_formatter(md.DateFormatter('%m/%d\n%H:%M', tz=pytz.timezone('Asia/Tokyo')))
ax.set_ylabel('Temperature [K]')
leg = ax.legend(loc='best')
leg.get_frame().set_alpha(0.5)
ax.grid()
plt.show()
```

`todlib.gentod()` を使うと、同様のプロットがより簡単に描けます。

```python
tod.gentod(10, ['ch. '+str(i) for i in range(8)]).plot()
plt.show()
```

`todlib.gentod()` は pandas の DataFrame を返すので、`plot()` に限らず、その他色々なメソッドを使うことができます（ref., <http://pandas.pydata.org/pandas-docs/dev/generated/pandas.DataFrame.html>）。

ある特定の期間をプロットするには以下のようにします。

```python
tod.gentod(colslabel=['ch. '+str(i) for i in range(8)])['2014-10-25-0310':'2014-10-25-0320'].plot()
show()
```

`TOD` クラスをイニシャライズしたときに引数にとったファイルと同じ階層にある他のファイル（同じデータ構造のファイル）とデータをつなぎあわせて、ひとつの大きい TOD を作りたいときは、`todlib.gentod_all()` を使います。

```python
todAll = tod.gentod_all(colslabel=['ch. '+str(i) for i in range(8)])
print todAll
```

`todlib.gentod_all()` も `todlib.gentod()` と同様に DataFrame 形式を返します。

```python
ax = todAll.resample('H')['2014-09-26':'2014-10-25'].plot()
ax.set_ylabel('Temperature [K]')
plt.show()
```
